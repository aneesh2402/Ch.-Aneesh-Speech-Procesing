```python
import librosa
import IPython.display as ipd 
import matplotlib.pyplot as plt
import numpy as np
```

# A1.Load the recorded speech file into your python workspace. Once loaded, plot the graph for the speech signal.You may use the below code from librosa asa reference.

```python
y, sr = librosa.load('lab1_recording.wav')
librosa.display.waveshow(y)
plt.title("Original Voice")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show()
```

# A2. Observe the length & magnitude range of the signalfrom theplot. Observing the plot, relate tothe spoken words and silence between the words.

```python
print("Signal length:",len(y),"samples")
print("Magnitude Range:",np.min(y),"to", np.max(y))
print("Sampling Rate:",sr, "Hz")
```

# A3.Take a small segment of the signal and play it.

# A4. Play around with your recorded speech signal for various segments. Understand the nature of the signal. Also observe with abruptly segmented speech, how the perception of that speech is affected.

```python
start_time = 0
end_time = 1
sr1 = 20000
segment1 = y[int(start_time * sr1):int(end_time * sr1)]
print("Segment 1")
ipd.Audio(segment1, rate=sr1)
```

```python
plt.figure(figsize=(10, 5))
librosa.display.waveshow(segment1, sr=sr)
plt.title("Segment 1")
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
```
